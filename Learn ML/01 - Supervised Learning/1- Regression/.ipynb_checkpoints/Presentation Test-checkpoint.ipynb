# Data 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(1,15,50) # number of hours
```

```python
Y = np.linspace(4,9,50)  + np.random.randn(50) # CGPA
```

```python
plt.style.use('seaborn')
plt.scatter(X,Y,marker='*',s=60)
plt.xlim(0,18)
plt.ylim(0,14)
plt.title('MicroProcessor Paper Analysis')
plt.xlabel('Average Number of Hours studied over the period of 5 days before Exam')
plt.ylabel('Student - CGPA')
```
